```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv('Crime_Agains_Women_In_India_2015-2016.csv')
df.head()
```

```python
xpos = np.arange(len(df['States/UTs']))
xpos
```

```python

plt.figure(figsize=(16,20))
plt.xticks(xpos,df['States/UTs'],rotation=90,fontsize=12,color='blue')
plt.title('[ Crime Against Women In India 2015 ]',fontsize=16,color='r')
plt.xlabel('- Indian States -',fontsize=12)
plt.ylabel('- Cases -',fontsize=12)
plt.bar(xpos-0.3,df['Assaults (molestation) - 2015'],width=0.4,color='y',label='Molestation')
plt.bar(xpos-0.1,df['Rape - 2015'],width=0.4,color='coral',label='Rape')
plt.bar(xpos+0.1,df['Murder (women) - 2015'],width=0.4,color='r',label='Murder')
plt.legend(fontsize=14)
```
