```python
dfMH = pd.read_csv('Maharashtra_City_Data-1.csv')
dfMH.head()
```

```python
xpos = np.arange(len(dfMH['name_of_city']))
xpos
```

```python
plt.figure(figsize=(16,15))
plt.xticks(xpos,dfMH['name_of_city'],rotation=90,fontsize=12,color='blue')
plt.title('[ Maharashtra Cities Female Population]',fontsize=16)
plt.xlabel('- City -',fontsize=12)
plt.ylabel('- Population -',fontsize=12)
plt.bar(xpos-0.2,dfMH['population_male'],width=0.4,color='r',label='Male Population')
plt.bar(xpos+0.2,dfMH['population_female'],width=0.4,color='g',label='Female Population')
plt.legend(loc=7)
plt.grid()
```

```python
plt.figure(figsize=(16,4))
plt.xticks(xpos,dfMH['name_of_city'],rotation=90,fontsize=11.5)
plt.title('[ Maharashtra Cities with Sex Ratio]',fontsize=14,color='darkred')
plt.xlabel('-  City  -',fontsize=11)
plt.ylabel(' -  Female per 1000 Male  - ',fontsize=12)
plt.plot(xpos,dfMH['sex_ratio'],'--.b', label='sex_ratio')
plt.plot(xpos,dfMH['child_sex_ratio'],'--.r',label='child sex_ratio')
plt.legend(loc=4)
plt.grid()
```

```python
plt.figure(figsize=(16,4))
plt.xticks(xpos,dfMH['name_of_city'],rotation=90,fontsize=11)
plt.title('[ Effective Literacy Rate of Maharashtra Cities ]',fontsize=14)
plt.xlabel('-  City  -',fontsize=12,color='teal')
plt.ylabel('-  Literacy rate  -',fontsize=12,color='teal')
plt.plot(xpos,dfMH['effective_literacy_rate_male'],'--.',color='hotpink',label='effective_literacy_rate_male')
plt.plot(xpos,dfMH['effective_literacy_rate_female'],'--.',color='indigo',label='effective_literacy_rate_female')
plt.legend(loc=3)
plt.grid()
```

```python
plt.figure(figsize=(16,13))
plt.xticks(xpos,dfMH['name_of_city'],rotation=90,fontsize=11)
plt.title('[ Graduates from Maharashtra Cities ]',fontsize=14)
plt.xlabel('-  City  -',fontsize=12,color='teal')
plt.ylabel('-  No of Graduates  -',fontsize=12,color='teal')
plt.plot(xpos,dfMH['male_graduates'],'--.',color='maroon',label='male_graduates')
plt.plot(xpos,dfMH['female_graduates'],'--.',color='lime',label='female_graduates')
plt.grid()
```
