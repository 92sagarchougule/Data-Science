```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('datafile.csv')
df.head()
```

```python
df.size
```

```python
df.shape
```

```python
xpos = np.arange(len(df['States/UTs']))
xpos
```

```python
plt.figure(figsize=(20,20))
plt.title('STATES WISE FOREST FIRE IN INDIA ')
plt.xlabel('Indian States',fontsize=14)
plt.ylabel('Fire Cases',fontsize=14)
plt.xticks(xpos,df['States/UTs'],rotation=90,fontsize=15,color='blue')
plt.bar(xpos+0.3,df['2010-2011'],width=0.2,color='r',label='2010-2011')
plt.bar(xpos+0.1,df['2009-10'],width=0.2,color='coral',label='2009-10')
plt.bar(xpos-0.1,df['2008-09'],width=0.2,color='y',label='2008-09')

plt.legend(fontsize=14)
```
